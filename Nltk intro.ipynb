# POS tagging & Chunking

```python
from nltk.tokenize import PunktSentenceTokenizer
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(abc.raw("rural.txt")).tokenize(sample)
```

```python
try:
    for i in custom_sent_tokenizer:
        wrd = word_tokenize(i)
        tagged = nltk.pos_tag(wrd)
        
        #print(tagged)
        
        chunkgram = r"""Chunk: {<NNP>*<NN>*} """
        chunkParser = nltk.RegexpParser(chunkgram)
        chunked = chunkParser.parse(tagged)
        
        chunked.draw()
except Exception as e:
    print(str(e))
```

# NER

```python
try:
    for i in custom_sent_tokenizer:
        wrd = word_tokenize(i)
        tagged = nltk.pos_tag(wrd)
        
        nameEnt = nltk.ne_chunk(tagged ,binary =True)
        nameEnt.draw()
        
except Exception as e:
    print(str(e))
```

```python
synonyms=[]
antonyms=[]
```

```python
for syn in wordnet.synsets('good'):
    for l in syn.lemmas():
        synonyms.append(l.name())
        
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())
```

```python
synonyms = set(synonyms)
synonyms
```
